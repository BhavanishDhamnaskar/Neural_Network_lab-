```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import make_classification

# Generate dataset
features, labels = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)

# Manually split dataset
indices = np.arange(features.shape[0])
np.random.shuffle(indices)
split_idx = int(0.8 * len(features))
train_idx, test_idx = indices[:split_idx], indices[split_idx:]
train_features, test_features = features[train_idx], features[test_idx]
train_labels, test_labels = labels[train_idx], labels[test_idx]

# Define and compile model
model = Sequential([
    Dense(10, activation='relu', input_shape=(features.shape[1],)),
    Dense(8, activation='relu'),
    Dense(8, activation='relu'),
    Dense(4, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Print the model summary to check the architecture
model.summary()

# Train model with verbose output to show progress for each epoch
history = model.fit(train_features, train_labels, epochs=100, validation_split=0.2, verbose=1)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import make_classification

# Generate dataset
features, labels = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)

# Manually split dataset using NumPy
np.random.seed(42)  # Ensure reproducibility
indices = np.arange(features.shape[0])
np.random.shuffle(indices)
split_idx = int(0.8 * len(features))  # 80% for training, 20% for testing

train_indices, test_indices = indices[:split_idx], indices[split_idx:]
train_features, test_features = features[train_indices], features[test_indices]
train_labels, test_labels = labels[train_indices], labels[test_indices]

# Define and compile model
nn_model = Sequential([
    Dense(10, activation='relu', input_shape=(20,)),  # Assuming 20 features as specified
    Dense(8, activation='relu'),
    Dense(8, activation='relu'),
    Dense(4, activation='relu'),
    Dense(1, activation='sigmoid')
])
nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Convert the data to a tf.data.Dataset and batch it
batch_size = 32
train_dataset = tf.data.Dataset.from_tensor_slices((train_features, train_labels)).shuffle(buffer_size=1024).batch(batch_size)
val_dataset = tf.data.Dataset.from_tensor_slices((test_features, test_labels)).batch(batch_size)

# Train model using tf.data.Dataset
history = nn_model.fit(train_dataset, epochs=100, validation_data=val_dataset)

# Evaluate model
test_loss, test_acc = nn_model.evaluate(test_features, test_labels, verbose=2)
print('\nTest accuracy:', test_acc)
```
